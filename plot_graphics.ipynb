```python
from scipy.interpolate import interp1d
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

# plt.plot(validacao["Re"], validacao["Texp"], '--')
# plt.plot(validacao["Re"], validacao["Tnum"], '--')

fexp = interp1d(validacao["Re"], validacao['Texp'], kind='cubic')
plt.plot(validacao["Re"], fexp(validacao["Re"]), '--')


fnum = interp1d(validacao["Re"], validacao['Tnum'], kind='cubic')
plt.plot(validacao["Re"], fnum(validacao["Re"]), '-')

# sns.lineplot(x="Re", y='Texp',data=validacao,ci=None,linestyle='--')
# sns.lineplot(x="Re", y='Tnum',data=validacao,ci=None,linestyle='-')
plt.legend(['Experimental (MASCHIETTO, 2018)', 'Presente Trabalho'])
plt.ylabel('Tmax')
plt.xlabel('Reynolds')

plt.show()
```

```python
from scipy.interpolate import interp1d
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline


plt.figure(figsize=(10,6))

# plt.plot(validacao["Re"], validacao["Texp"], '--')
# plt.plot(validacao["Re"], validacao["Tnum"], '--')


fexp = UnivariateSpline(validacao["Re"], validacao['Texp'], k = 3)
# fexp = interp1d(validacao["Re"], validacao['Texp'], kind='cubic')
plt.plot(validacao["Re"], fexp(validacao["Re"]), '--')


fnum = UnivariateSpline(validacao["Re"], validacao['Tnum'], k = 3)
plt.plot(validacao["Re"], fnum(validacao["Re"]), '-')

# sns.lineplot(x="Re", y='Texp',data=validacao,ci=None,linestyle='--')
# sns.lineplot(x="Re", y='Tnum',data=validacao,ci=None,linestyle='-')
plt.legend(['Experimental (MASCHIETTO, 2018)', 'Presente Trabalho'])
plt.ylabel('Tmax')
plt.xlabel('Reynolds')

plt.show()
```

```python
from scipy.interpolate import interp1d
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline
from scipy.interpolate import make_interp_spline
from scipy.ndimage import gaussian_filter1d

plt.figure(figsize=(10,6))

# plt.plot(validacao["Re"], validacao["Texp"], '--')
# plt.plot(validacao["Re"], validacao["Tnum"], '--')


fexp = make_interp_spline(validacao["Re"], validacao['Texp'])
# fexp = interp1d(validacao["Re"], validacao['Texp'], kind='cubic')
plt.plot(validacao["Re"], gaussian_filter1d(validacao['Texp'], sigma=1), '--')


fnum = make_interp_spline(validacao["Re"], validacao['Tnum'])
plt.plot(validacao["Re"], gaussian_filter1d(validacao['Tnum'], sigma=1), '-')

# sns.lineplot(x="Re", y='Texp',data=validacao,ci=None,linestyle='--')
# sns.lineplot(x="Re", y='Tnum',data=validacao,ci=None,linestyle='-')
plt.legend(['Experimental (MASCHIETTO, 2018)', 'Presente Trabalho'])
plt.ylabel('Tmax')
plt.xlabel('Reynolds')

plt.show()
```
