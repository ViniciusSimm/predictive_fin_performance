```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
sns.lineplot(x="Re", y='Texp',data=validacao,ci=None)
sns.lineplot(x="Re", y='Tnum',data=validacao,ci=None)
plt.legend(['Texp', 'Tnum'])
plt.ylabel('Tmax')
plt.xlabel('Reynolds')
```
